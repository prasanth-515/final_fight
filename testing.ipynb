```python
# -------------------------------
# Everyday Git Commands
# -------------------------------
# git init                -> Start a new Git repository
# git clone <url>         -> Copy a remote repo to local machine
# git status              -> Show changes (staged, unstaged, untracked)
# git add <file>          -> Stage a file for commit
# git commit -m "msg"     -> Save staged changes with a message
# git log --oneline       -> Show commit history (short form)
# git push origin master  -> Upload commits to GitHub
# git pull origin master  -> Download latest changes from GitHub
# git branch              -> List branches
# git checkout <branch>   -> Switch to another branch
# git merge <branch>      -> Merge another branch into current

# -------------------------------
# Fixing Mistakes
# -------------------------------
# git commit --amend -m "new msg"   -> Change last commit message
# git reset --soft HEAD~1           -> Undo last commit, keep changes staged
# git reset --hard HEAD~1           -> Undo last commit, discard changes
# git revert <hash>                 -> Create a new commit that undoes a past commit
# git rebase -i HEAD~N              -> Rewrite history, drop or edit commits
# git reflog                        -> Show all actions (recover lost commits)
# git cherry-pick <hash>            -> Reapply a commit from history

# -------------------------------
# Best Practices
# -------------------------------
# - Commit often with clear messages
# - Use revert for safety (keeps history intact)
# - Use rebase/drop only if you want to erase commits completely (requires force push)
# - Check git log or git reflog before panicking — almost nothing is truly lost
```

```python
# -------------------------------
# Git Branch Command 
# -------------------------------

# List all local branches
# git branch
# (Shows which branches exist locally, * marks the current branch)

# Create a new branch from current HEAD
# git checkout -b <branch_name>
# (Makes a new branch starting at your current commit)

# Create a branch from a specific commit
# git checkout -b <branch_name> <commit_hash>
# (Useful for making a sub-branch from an earlier commit, e.g. Git_Commands)

# Create a branch from another branch
# git checkout <other_branch>
# git checkout -b <new_branch>
# (Switch to the branch you want to base from, then create a new one)

# Switch to an existing branch
# git checkout <branch_name>
# (Moves HEAD to that branch)

# Push a new branch to GitHub
# git push origin <branch_name>
# (Publishes the branch so it’s visible remotely)

# Delete a branch locally (safe, only if merged)
# git branch -d <branch_name>

# Force delete a branch locally (even if not merged)
# git branch -D <branch_name>

# Delete a branch on GitHub (remote)
# git push origin --delete <branch_name>

# Rename a branch
# git branch -m <old_name> <new_name>
# (Renames the branch you’re on or a specific branch)

# See branch history with graph
# git log --oneline --graph --decorate --all
# (Shows commits with ASCII graph, branches, and tags)

# Set upstream (link local branch to remote branch)
# git branch --set-upstream-to=origin/<branch_name>
# (Ensures pushes/pulls go to the right remote branch)
```
